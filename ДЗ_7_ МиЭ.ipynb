Задача 4. 

Пусть CF1 = · · · = CF50 = 1000 — одинаковые задолженности от
большого количества (50) контрагентов. 

Пусть ξi — случайная величина,означающая дефолт i-го контрагента. 

Пусть есть общий фактор для всех контрагентов, т.е. ξi = η + ζi, где P(η = 1) = P(ζi = 1) = 0.01,(ζ1, . . . , ζ50)
— независимые случайные величины, P(η = 1, ζi = 1) = 0.

R =?; V aR1%(X) =?; V aR5%(X) =?; ESF1%(X) =?; ESF5%(X) =?.

```python
n=50
p1 = 0.01
b = []
for k in range(1, 51):
    i = binom.pmf(k,n,p1)
    b.append(i)
    print(f"Вероятность наступления дефолта {k} из {n} равна {poisson.pmf(n,p1)} {sum(b)}")
```
